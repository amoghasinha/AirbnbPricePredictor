```python
import numpy as np 

def iqr(df, var):
    q1 = np.quantile(df[var], 0.25)
    q3 = np.quantile(df[var], 0.75)
    diff = q3-q1
    lower_v = q1 - (1.5*diff)
    upper_v = q3 + (1.5*diff)
    return df[(df[var]<lower_v) | (df[var]>upper_v)]
```

```python
# Now, we move onto hypothesis testing
# I want to test whether the mean price for private rooms are different than the mean price for entire homes/apartments.

# Null Hypothesis (H0): There is no significant difference in average price between "Entire home/apartment" and 
# "Private room"
# Alternative Hypothesis (H1): There is a significant difference in average price between "Entire home/apartment" and 
# "Private room."

entire_home_data = df[df['room_type'] == 'Entire home/apt']
private_room_data = df[df['room_type'] == 'Private room']

mean1 = entire_home_data['price'].mean()
std1 = entire_home_data['price'].std()

mean2 = private_room_data['price'].mean()
std2 = private_room_data['price'].std()

print("Mean and sd of Entire Home data:", mean1, std1)
print("Mean and sd of Private Room data: ", mean2, std2)
```

```python
import math

n1 = len(entire_home_data)
n2 = len(private_room_data)

se = math.sqrt((std1**2/n1) + (std2**2/n2))
print(se)
```

```python
from scipy.stats import norm
z = (mean1 - mean2) / se
print(z)
p_value = 2 * (1 - norm.cdf(abs(z)))
print(p_value)
```

```python
# Because the p-value is 0.0, we reject the null hypothesis. There is a significant difference between the price of 
# Entire Home airbnbs and Private Room airbnbs
```

```python
import pandas as pd
from scipy.stats import f_oneway

result = f_oneway(*[df['price'][df['neighbourhood_group'] == group] for group in df['neighbourhood_group'].unique()])

print("ANOVA p-value:", result.pvalue)

# Check if the p-value is less than the significance level (e.g., 0.05) to determine significance
if result.pvalue < 0.05:
    print("There is a significant difference in price between the neighborhood groups.")
else:
    print("There is no significant difference in price between the neighborhood groups.")
```

```python
latitude = [40.7116, 40.7794, 40.7826, 40.7484, 40.8296, 40.6892, 40.7587, 40.7593, 40.7127, 40.7580, 40.7061, 40.7505, 
            40.7534, 40.7473, 40.7908, 40.6446]
longitude = [-74.0133, -73.9632, -73.9656, -73.9857, -73.9262, -74.0445, -73.9787, -73.9794, -74.0134 , -73.9855, -73.9969, -73.9934,
           -73.9768, -73.8517, -73.9747, -73.7797]

location_df = location_df.dropna(how='all')
print(location_df)

location_df.loc[:, 'latitude'] = latitude
location_df.loc[:, 'longitude'] = longitude
location_df
```

```python
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = 6371 * c  # Radius of Earth in kilometers

    return distance

df['within_1_mile'] = False  # Initialize as False

# Iterate through each row in 'airbnb_df'
for idx, airbnb_row in df.iterrows():
    airbnb_lat, airbnb_lon = airbnb_row['latitude'], airbnb_row['longitude']

    for att_idx, att_row in location_df.iterrows():
        att_lat, att_lon = att_row['latitude'], att_row['longitude']

        # Calculate distance using Haversine formula
        distance = haversine(airbnb_lat, airbnb_lon, att_lat, att_lon)

        # If within 1 mile, set 'within_1_mile' to True and break the loop
        if distance <= 1.0:
            df.at[idx, 'within_1_mile'] = True
            break

# Display the resulting DataFrame
print(df.head())
```
